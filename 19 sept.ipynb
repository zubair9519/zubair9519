```
import requests

# Your organization ID and API token
organization_id = '      '
api_token = '    '

# API endpoint to get circles in the organization
api_url = f'https://app.holaspirit.com/api/organizations/{organization_id}/circles?page=1&count=30'

# Headers for the GET request
headers = {
    'Authorization': f'Bearer {api_token}',
    'Accept': 'application/json'
}

# Send the GET request
response = requests.get(api_url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    print("Data successfully retrieved!")
    data = response.json()

    # Create a list to store the parent and sub-circles
    parent_and_sub_circles = []

    # Loop through the circles data to extract parentCircle and subRoles
    for circle in data['data']:
        parent_circle = circle.get('parentCircle', 'None')
        sub_circles = circle.get('subRoles', [])

        # Append the parent circle and sub-circle information
        parent_and_sub_circles.append({
            'Parent Circle': parent_circle,
            'Sub Circles': sub_circles
        })

    # Display the list of parent circles and their sub-circles
    for entry in parent_and_sub_circles:
        print(f"Parent Circle ID: {entry['Parent Circle']}")
        print(f"Sub Circles: {entry['Sub Circles']}")
        print("-" * 40)

else:
    print(f"Failed to access the API. Status code: {response.status_code}")
    print("Error message:", response.text)
```

```
# Assuming 'data' contains the response from the API
# Let's focus on filtering and displaying parent circles and their sub-circles.

# Create an empty dictionary to store parent circles and their sub-circles
parent_circles_and_subcircles = {}

# Loop through each circle in the data
for circle in data['data']:
    # Get the parent circle ID
    parent_circle = circle.get('parentCircle', None)

    # Get the current circle ID
    current_circle_id = circle.get('id', None)

    # Get the list of subRoles (sub-circles)
    sub_roles = circle.get('subRoles', [])

    # If the circle has a parent, we'll consider it a parent circle
    if parent_circle:
        if parent_circle not in parent_circles_and_subcircles:
            # Add the parent circle to the dictionary if not already added
            parent_circles_and_subcircles[parent_circle] = []

        # Add the current circle as a sub-circle to its parent
        parent_circles_and_subcircles[parent_circle].append(current_circle_id)

# Now print the parent circles and their sub-circles
for parent_circle, sub_circles in parent_circles_and_subcircles.items():
    print(f"Parent Circle ID: {parent_circle}")
    print(f"Sub-Circles: {sub_circles}")
    print("-" * 40)
```

```
# Create a dictionary to map role IDs to their accountabilities and assigned members
role_id_to_accountabilities = {role['id']: role.get('purpose', 'No Accountability Provided') for role in data['linked']['roles']}
role_id_to_assigned_members = {role['id']: role.get('members', ['No Assigned Members']) for role in data['linked']['roles']}

# Iterate through each circle and its roles
for circle_name, info in circle_roles_with_parents.items():
    print(f"Circle Name: {circle_name}")
    print(f"Circle ID: {info['circle_id']}")
    print(f"Parent Circle Name: {info['parent_circle_name']}")

    # Iterate through roles in the circle
    for role_name in info['roles']:
        # Find the role ID by its name
        role_id = next((role['id'] for role in data['linked']['roles'] if role['name'] == role_name), None)

        # Get accountabilities and assigned members if role_id is found
        if role_id:
            accountabilities = role_id_to_accountabilities.get(role_id, 'No Accountability Provided')
            assigned_members = role_id_to_assigned_members.get(role_id, ['No Assigned Members'])

            # Safely retrieve member names by checking if the item is a dictionary
            member_names = [
                member.get('name', 'Unknown Member') if isinstance(member, dict) else member
                for member in assigned_members
            ]
        else:
            accountabilities = 'No Accountability Found'
            member_names = ['No Assigned Members']

        print(f"Role: {role_name}")
        print(f"Accountabilities: {accountabilities}")
        print(f"Assigned Members: {', '.join(member_names)}")
        print("-" * 40)

    print("=" * 60)
```
